```python
## Read original metadata
xls = pd.ExcelFile('../Study 1/Unblinded/Submission Review Report_BIMO_Listing_Study1_Unblinded.xlsx')
df_orig = pd.read_excel(xls, sheet_name='original_metadata',header=None )
df_orig[:10]
```

```python
df_orig2 = pd.DataFrame(columns=['index', 'Variable', 'Label','Type','CT','desc','sample'])
df_orig2
```

```python
for i, _df1 in df_orig.iterrows():
    rem = i % 7
    div = int(i / 7) + 1 
    if rem == 0:
        #print("the remainder is", i, rem, div)
        #df_orig2.iloc[div, 'index'] = _df1
        df_orig2.loc[div, 'index'] = _df1[0]
    if rem == 1:
        df_orig2.loc[div, 'Variable'] = _df1[0]
    if rem == 2:
        df_orig2.loc[div, 'Label'] = _df1[0]
    if rem == 3:
        df_orig2.loc[div, 'Type'] = _df1[0]
    if rem == 4:
        df_orig2.loc[div, 'CT'] = _df1[0]
    if rem == 5:
        df_orig2.loc[div, 'desc'] = _df1[0]
    if rem == 6:
        df_orig2.loc[div, 'sample'] = _df1[0]
df_orig2
```

```python
df_orig3 = df_orig2[['Variable','Type','Label']]
df_orig3['Type'] = df_orig3.Type.replace(['Char','Num'],['Character','Numeric']) 
df_orig3.head()
```

```python
df_clin = pd.read_excel(xls, sheet_name='clinsite_metadata')
df_clin.head()
```

```python
df_clin2 = df_clin[['Variable','Type','Label']]
df_clin2.head()
```
